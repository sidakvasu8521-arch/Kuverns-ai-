```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
import warnings
warnings.filterwarnings('ignore')

# Data processing
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.impute import SimpleImputer

# Clustering
from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn.metrics import silhouette_score, davies_bouldin_score

# Feature engineering
from scipy import stats
from sklearn.decomposition import PCA

# Modeling
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Visualization
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# Set random seed for reproducibility
np.random.seed(42)

# Configure display settings
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 100)
plt.style.use('seaborn-v0_8-darkgrid')

print("✅ All libraries imported successfully!")
print(f"Execution started at: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")

# ==============================================================================
# DATA LOADING AND INITIAL EXPLORATION
# ==============================================================================

class DataLoader:
    """Class to handle data loading and initial processing"""

    def __init__(self):
        self.data_url = "https://archive.ics.uci.edu/ml/machine-learning-databases/00235/household_power_consumption.zip"
        self.df = None

    def load_data(self):
        """Load the household power consumption dataset"""
        print("\n📊 Loading Household Power Consumption Dataset...")
        print(f"Source: {self.data_url}")

        try:
            # Download and extract the dataset
            import zipfile
            import requests
            from io import BytesIO

            # Download the zip file
            response = requests.get(self.data_url)
            zip_file = zipfile.ZipFile(BytesIO(response.content))

            # Extract and read the data
            for file_name in zip_file.namelist():
                if file_name.endswith('.txt'):
                    with zip_file.open(file_name) as file:
                        self.df = pd.read_csv(file, sep=';', low_memory=False)
                    break

            print(f"✅ Data loaded successfully! Shape: {self.df.shape}")
            return self.df

        except Exception as e:
            print(f"⚠️ Error loading from URL: {e}")
            print("Loading alternative sample data...")
            return self.load_sample_data()

    def load_sample_data(self):
        """Create sample data if original dataset fails to load"""
        np.random.seed(42)
        n_samples = 10000

        dates = pd.date_range(start='2020-01-01', periods=n_samples, freq='H')

        self.df = pd.DataFrame({
            'Date': dates.date,
            'Time': dates.time,
            'Global_active_power': np.random.normal(1.5, 0.5, n_samples),
            'Global_reactive_power': np.random.normal(0.3, 0.1, n_samples),
            'Voltage': np.random.normal(240, 5, n_samples),
            'Global_intensity': np.random.normal(6, 2, n_samples),
            'Sub_metering_1': np.random.exponential(2, n_samples),
            'Sub_metering_2': np.random.exponential(1.5, n_samples),
            'Sub_metering_3': np.random.exponential(3, n_samples)
        })

        # Add some patterns
        self.df['Global_active_power'] += np.sin(np.arange(n_samples) * 2 * np.pi / 24) * 0.3

        print(f"✅ Sample data created! Shape: {self.df.shape}")
        return self.df

# ==============================================================================
# DATA PREPROCESSING
# ==============================================================================

class DataPreprocessor:
    """Class to handle all data preprocessing steps"""

    def __init__(self, df):
        self.df = df.copy()

    def preprocess(self):
        """Main preprocessing pipeline"""
        print("\n🔧 Starting Data Preprocessing...")

        # 1. Handle datetime
        self.handle_datetime()

        # 2. Handle missing values
        self.handle_missing_values()

        # 3. Convert data types
        self.convert_datatypes()

        # 4. Remove outliers
        self.remove_outliers()

        # 5. Sort by datetime
        self.df = self.df.sort_values('datetime').reset_index(drop=True)

        print("✅ Preprocessing completed!")
        return self.df

    def handle_datetime(self):
        """Combine date and time columns into datetime"""
        try:
            self.df['datetime'] = pd.to_datetime(
                self.df['Date'].astype(str) + ' ' + self.df['Time'].astype(str),
                format='%d/%m/%Y %H:%M:%S',
                errors='coerce'
            )
        except:
            # Alternative format
            self.df['datetime'] = pd.to_datetime(
                self.df['Date'].astype(str) + ' ' + self.df['Time'].astype(str),
                errors='coerce'
            )

        # Drop rows with invalid datetime
        self.df = self.df.dropna(subset=['datetime'])
        print(f"  ✓ Datetime created. Valid rows: {len(self.df)}")

    def handle_missing_values(self):
        """Handle missing values in the dataset"""
        # Replace '?' with NaN
        self.df = self.df.replace('?', np.nan)

        # Check missing values
        missing = self.df.isnull().sum()
        print(f"  ✓ Missing values per column:\n{missing[missing > 0]}")

        # Impute numerical columns
        numerical_cols = self.df.select_dtypes(include=[np.number]).columns
        if len(numerical_cols) > 0:
            imputer = SimpleImputer(strategy='mean')
            self.df[numerical_cols] = imputer.fit_transform(self.df[numerical_cols])

    def convert_datatypes(self):
        """Convert columns to appropriate data types"""
        numeric_columns = [
            'Global_active_power', 'Global_reactive_power', 'Voltage',
            'Global_intensity', 'Sub_metering_1', 'Sub_metering_2', 'Sub_metering_3'
        ]

        for col in numeric_columns:
            if col in self.df.columns:
                self.df[col] = pd.to_numeric(self.df[col], errors='coerce')

        print(f"  ✓ Data types converted")

    def remove_outliers(self, threshold=3):
        """Remove outliers using Z-score method"""
        numeric_cols = self.df.select_dtypes(include=[np.number]).columns

        initial_rows = len(self.df)
        for col in numeric_cols:
            # Calculate z_scores for the entire column. scipy.stats.zscore handles NaNs by returning NaN at those positions.
            z_scores = np.abs(stats.zscore(self.df[col]))

            # Create a boolean mask: keep rows where z_score is below threshold OR
            # where the original value in 'col' was NaN (and thus z_score is also NaN).
            # This correctly handles rows that were originally NaN and should not be removed as outliers based on non-existent values.
            mask = (z_scores < threshold) | self.df[col].isna()
            self.df = self.df[mask]

        print(f"  ✓ Outliers removed. Remaining rows: {len(self.df)} (Removed {initial_rows - len(self.df)} outliers)")

# ==============================================================================
# FEATURE ENGINEERING
# ==============================================================================

class FeatureEngineer:
    """Class to create features for clustering and modeling"""

    def __init__(self, df):
        self.df = df.copy()

    def create_temporal_features(self):
        """Create time-based features"""
        self.df['hour'] = self.df['datetime'].dt.hour
        self.df['day_of_week'] = self.df['datetime'].dt.dayofweek
        self.df['month'] = self.df['datetime'].dt.month
        self.df['is_weekend'] = (self.df['day_of_week'] >= 5).astype(int)

        # Time of day categories
        self.df['time_of_day'] = pd.cut(
            self.df['hour'],
            bins=[0, 6, 12, 18, 24],
            labels=['Night', 'Morning', 'Afternoon', 'Evening']
        )

        print("  ✓ Temporal features created")
        return self.df

    def create_aggregated_features(self):
        """Create aggregated features for household clustering"""
        # Assuming we need to aggregate by day for household patterns
        daily_features = self.df.groupby(self.df['datetime'].dt.date.rename('date')).agg({
            'Global_active_power': ['mean', 'max', 'min', 'std'],
            'Global_reactive_power': ['mean', 'max'],
            'Voltage': ['mean', 'std'],
            'Global_intensity': ['mean', 'max'],
            'Sub_metering_1': 'sum',
            'Sub_metering_2': 'sum',
            'Sub_metering_3': 'sum'
        })

        # Flatten column names
        daily_features.columns = ['_'.join(col).strip() for col in daily_features.columns.values]
        daily_features = daily_features.reset_index()

        # Additional features
        daily_features['total_sub_metering'] = (
            daily_features['Sub_metering_1_sum'] +
            daily_features['Sub_metering_2_sum'] +
            daily_features['Sub_metering_3_sum']
        )

        # Peak hour identification (requires hourly data)
        hourly_power = self.df.groupby([
            self.df['datetime'].dt.date.rename('date'),
            self.df['datetime'].dt.hour.rename('hour')
        ])['Global_active_power'].mean().reset_index()

        peak_hours = hourly_power.loc[
            hourly_power.groupby('date')['Global_active_power'].idxmax()
        ]
        peak_hours.columns = ['date', 'peak_hour', 'peak_power']

        # Merge peak hour info
        daily_features = pd.merge(daily_features, peak_hours, on='date', how='left')

        print(f"  ✓ Aggregated features created. Shape: {daily_features.shape}")
        return daily_features

    def create_lag_features(self, n_lags=3):
        """Create lag features for time series forecasting"""
        for lag in range(1, n_lags + 1):
            self.df[f'power_lag_{lag}'] = self.df['Global_active_power'].shift(lag)

        # Rolling statistics
        for window in [3, 7, 24]:
            self.df[f'power_roll_mean_{window}'] = (
                self.df['Global_active_power'].rolling(window=window, min_periods=1).mean()
            )
            self.df[f'power_roll_std_{window}'] = (
                self.df['Global_active_power'].rolling(window=window, min_periods=1).std()
            )

        print(f"  ✓ Lag features created with {n_lags} lags")
        return self.df

# ==============================================================================
# CLUSTERING ANALYSIS
# ==============================================================================

class ClusteringAnalysis:
    """Class to perform clustering analysis on households"""

    def __init__(self, features_df):
        self.features_df = features_df.copy()
        self.scaler = StandardScaler()
        self.clusters = None
        self.best_k = None

    def prepare_clustering_data(self):
        """Prepare data for clustering"""
        # Select features for clustering
        clustering_features = [
            'Global_active_power_mean', 'Global_active_power_max',
            'Global_active_power_std', 'total_sub_metering',
            'peak_hour', 'peak_power'
        ]

        # Check which features exist
        available_features = [f for f in clustering_features if f in self.features_df.columns]

        # Handle missing values
        self.X_clustering = self.features_df[available_features].fillna(
            self.features_df[available_features].mean()
        )

        # Scale features
        self.X_clustering_scaled = self.scaler.fit_transform(self.X_clustering)

        print(f"  ✓ Clustering data prepared. Features: {len(available_features)}")
        return self.X_clustering_scaled

    def find_optimal_clusters(self, max_k=10):
        """Find optimal number of clusters using elbow method and silhouette score"""
        X = self.prepare_clustering_data()

        inertias = []
        silhouette_scores = []

        for k in range(2, min(max_k + 1, len(X))):
            kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
            labels = kmeans.fit_predict(X)

            inertias.append(kmeans.inertia_)
            silhouette_scores.append(silhouette_score(X, labels))

        # Plot results
        fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

        ax1.plot(range(2, min(max_k + 1, len(X))), inertias, 'bo-')
        ax1.set_xlabel('Number of Clusters')
        ax1.set_ylabel('Inertia')
        ax1.set_title('Elbow Method')
        ax1.grid(True)

        ax2.plot(range(2, min(max_k + 1, len(X))), silhouette_scores, 'ro-')
        ax2.set_xlabel('Number of Clusters')
        ax2.set_ylabel('Silhouette Score')
        ax2.set_title('Silhouette Analysis')
        ax2.grid(True)

        plt.tight_layout()
        plt.show()

        # Select best k based on silhouette score
        self.best_k = silhouette_scores.index(max(silhouette_scores)) + 2
        print(f"  ✓ Optimal number of clusters: {self.best_k}")

        return self.best_k

    def perform_kmeans_clustering(self, n_clusters=None):
        """Perform K-Means clustering"""
        if n_clusters is None:
            n_clusters = self.best_k if self.best_k else 3

        X = self.X_clustering_scaled

        # K-Means clustering
        kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init=10)
        self.clusters = kmeans.fit_predict(X)

        # Add cluster labels to features dataframe
        self.features_df['cluster_kmeans'] = self.clusters

        # Calculate metrics
        silhouette = silhouette_score(X, self.clusters)
        davies_bouldin = davies_bouldin_score(X, self.clusters)

        print(f"  ✓ K-Means clustering completed")
        print(f"    - Silhouette Score: {silhouette:.3f}")
        print(f"    - Davies-Bouldin Score: {davies_bouldin:.3f}")

        return self.clusters

    def perform_hierarchical_clustering(self, n_clusters=None):
        """Perform Hierarchical clustering"""
        if n_clusters is None:
            n_clusters = self.best_k if self.best_k else 3

        X = self.X_clustering_scaled

        # Hierarchical clustering
        hierarchical = AgglomerativeClustering(n_clusters=n_clusters)
        clusters_hierarchical = hierarchical.fit_predict(X)

        # Add cluster labels to features dataframe
        self.features_df['cluster_hierarchical'] = clusters_hierarchical

        # Calculate metrics
        silhouette = silhouette_score(X, clusters_hierarchical)

        print(f"  ✓ Hierarchical clustering completed")
        print(f"    - Silhouette Score: {silhouette:.3f}")

        return clusters_hierarchical

    def visualize_clusters(self):
        """Visualize clustering results"""
        # PCA for visualization
        pca = PCA(n_components=2)
        X_pca = pca.fit_transform(self.X_clustering_scaled)

        # Create visualization dataframe
        viz_df = pd.DataFrame({
            'PC1': X_pca[:, 0],
            'PC2': X_pca[:, 1],
            'Cluster': self.clusters.astype(str)
        })

        # Plotly scatter plot
        fig = px.scatter(
            viz_df, x='PC1', y='PC2', color='Cluster',
            title='Household Clusters (PCA Visualization)',
            labels={'PC1': f'PC1 ({pca.explained_variance_ratio_[0]:.2%} variance)',
                   'PC2': f'PC2 ({pca.explained_variance_ratio_[1]:.2%} variance)'}
        )

        fig.update_layout(height=500, width=700)
        fig.show()

# ==============================================================================
# FORECASTING MODELS
# ==============================================================================

class ForecastingModels:
    """Class to handle all forecasting models"""

    def __init__(self, df, features_df):
        self.df = df.copy()
        self.features_df = features_df.copy()
        self.models = {}
        self.results = {}

    def prepare_forecasting_data(self):
        """Prepare data for forecasting"""
        # Create feature engineer
        fe = FeatureEngineer(self.df)

        # Create temporal and lag features
        self.df = fe.create_temporal_features()
        self.df = fe.create_lag_features(n_lags=24)

        # Remove rows with NaN in lag features
        self.df = self.df.dropna()

        # Define features for modeling
        feature_cols = [
            'hour', 'day_of_week', 'month', 'is_weekend',
            'Global_reactive_power', 'Voltage', 'Global_intensity'
        ]

        # Add lag features
        lag_cols = [col for col in self.df.columns if 'lag' in col or 'roll' in col]
        feature_cols.extend(lag_cols)

        # Filter available features
        feature_cols = [col for col in feature_cols if col in self.df.columns]

        # Prepare X and y
        self.X = self.df[feature_cols].values
        self.y = self.df['Global_active_power'].values

        print(f"  ✓ Forecasting data prepared. Shape: X={self.X.shape}, y={self.y.shape}")
        return self.X, self.y

    def split_data(self, test_size=0.2, val_size=0.1):
        """Split data into train, validation, and test sets"""
        # First split: train+val and test
        X_temp, self.X_test, y_temp, self.y_test = train_test_split(
            self.X, self.y, test_size=test_size, shuffle=False
        )

        # Second split: train and val
        val_size_adjusted = val_size / (1 - test_size)
        self.X_train, self.X_val, self.y_train, self.y_val = train_test_split(
            X_temp, y_temp, test_size=val_size_adjusted, shuffle=False
        )

        print(f"  ✓ Data split completed:")
        print(f"    - Train: {self.X_train.shape}")
        print(f"    - Val: {self.X_val.shape}")
        print(f"    - Test: {self.X_test.shape}")

        return self.X_train, self.X_val, self.X_test, self.y_train, self.y_val, self.y_test

    def train_random_forest(self, **kwargs):
        """Train Random Forest Regressor"""
        print("\n🌲 Training Random Forest Regressor...")

        # Default parameters
        params = {
            'n_estimators': 100,
            'max_depth': 10,
            'min_samples_split': 5,
            'min_samples_leaf': 2,
            'random_state': 42,
            'n_jobs': -1
        }
        params.update(kwargs)

        # Train model
        rf_model = RandomForestRegressor(**params)
        rf_model.fit(self.X_train, self.y_train)

        # Predictions
        y_pred_train = rf_model.predict(self.X_train)
        y_pred_val = rf_model.predict(self.X_val)
        y_pred_test = rf_model.predict(self.X_test)

        # Store model and results
        self.models['RandomForest'] = rf_model
        self.results['RandomForest'] = {
            'train': self.evaluate_predictions(self.y_train, y_pred_train),
            'val': self.evaluate_predictions(self.y_val, y_pred_val),
            'test': self.evaluate_predictions(self.y_test, y_pred_test),
            'predictions': y_pred_test
        }

        print("  ✓ Random Forest training completed")
        self.print_model_results('RandomForest')

        return rf_model

    def train_gradient_boosting(self, **kwargs):
        """Train Gradient Boosting Regressor"""
        print("\n🚀 Training Gradient Boosting Regressor...")

        # Default parameters
        params = {
            'n_estimators': 100,
            'learning_rate': 0.1,
            'max_depth': 5,
            'min_samples_split': 5,
            'min_samples_leaf': 2,
            'random_state': 42,
            'subsample': 0.8
        }
        params.update(kwargs)

        # Train model
        gb_model = GradientBoostingRegressor(**params)
        gb_model.fit(self.X_train, self.y_train)

        # Predictions
        y_pred_train = gb_model.predict(self.X_train)
        y_pred_val = gb_model.predict(self.X_val)
        y_pred_test = gb_model.predict(self.X_test)

        # Store model and results
        self.models['GradientBoosting'] = gb_model
        self.results['GradientBoosting'] = {
            'train': self.evaluate_predictions(self.y_train, y_pred_train),
            'val': self.evaluate_predictions(self.y_val, y_pred_val),
            'test': self.evaluate_predictions(self.y_test, y_pred_test),
            'predictions': y_pred_test
        }

        print("  ✓ Gradient Boosting training completed")
        self.print_model_results('GradientBoosting')

        return gb_model

    def evaluate_predictions(self, y_true, y_pred):
        """Calculate evaluation metrics"""
        return {
            'rmse': np.sqrt(mean_squared_error(y_true, y_pred)),
            'mae': mean_absolute_error(y_true, y_pred),
            'r2': r2_score(y_true, y_pred)
        }

    def print_model_results(self, model_name):
        """Print model evaluation results"""
        results = self.results[model_name]
        print(f"\n  {model_name} Results:")
        print(f"  {'='*50}")
        print(f"  | Dataset | RMSE    | MAE     | R² Score |")
        print(f"  |---------|---------|---------|----------|")
        for dataset in ['train', 'val', 'test']:
            metrics = results[dataset]
            print(f"  | {dataset:7} | {metrics['rmse']:7.4f} | {metrics['mae']:7.4f} | {metrics['r2']:8.4f} |")
        print(f"  {'='*50}")

    def compare_models(self):
        """Compare all trained models"""
        print("\n📊 Model Comparison")
        print("="*70)

        comparison_df = pd.DataFrame()
        for model_name, results in self.results.items():
            for dataset in ['train', 'val', 'test']:
                for metric, value in results[dataset].items():
                    comparison_df.loc[model_name, f"{dataset}_{metric}"] = value

        print(comparison_df.round(4))

        # Visualize comparison
        fig, axes = plt.subplots(1, 3, figsize=(15, 5))

        metrics = ['rmse', 'mae', 'r2']
        titles = ['RMSE (Lower is Better)', 'MAE (Lower is Better)', 'R² Score (Higher is Better)']

        for idx, (metric, title) in enumerate(zip(metrics, titles)):
            test_values = [self.results[model]['test'][metric] for model in self.results.keys()]
            axes[idx].bar(self.results.keys(), test_values)
            axes[idx].set_title(title)
            axes[idx].set_ylabel(metric.upper())
            axes[idx].grid(True, alpha=0.3)

        plt.tight_layout()
        plt.show()

        return comparison_df

# ==============================================================================
# CLUSTER-BASED FORECASTING
# ==============================================================================

class ClusterBasedForecasting:
    """Train separate models for each cluster"""

    def __init__(self, df, features_df):
        self.df = df.copy()
        self.features_df = features_df.copy()
        self.cluster_models = {}
        self.cluster_results = {}

    def train_cluster_models(self, cluster_col='cluster_kmeans'):
        """Train separate models for each cluster"""
        print("\n🎯 Training Cluster-Based Models...")

        # Merge cluster information with main dataframe
        # For simplicity, we'll assign clusters based on date patterns
        unique_clusters = self.features_df[cluster_col].unique()

        for cluster_id in unique_clusters:
            print(f"\n  Training models for Cluster {cluster_id}...")

            # Filter data for this cluster
            cluster_dates = self.features_df[
                self.features_df[cluster_col] == cluster_id
            ]['date'].values

            # Create cluster-specific dataset
            cluster_df = self.df[
                self.df['datetime'].dt.date.isin(cluster_dates)
            ].copy()

            if len(cluster_df) > 100:  # Minimum samples required
                # Initialize forecasting models for this cluster
                fm = ForecastingModels(cluster_df, self.features_df)

                # Prepare data
                fm.prepare_forecasting_data()
                fm.split_data()

                # Train models
                fm.train_random_forest(n_estimators=50)
                fm.train_gradient_boosting(n_estimators=50)

                # Store results
                self.cluster_models[cluster_id] = fm.models
                self.cluster_results[cluster_id] = fm.results

        print("\n✅ Cluster-based model training completed!")

    def compare_cluster_vs_global(self, global_results):
        """Compare cluster-based models with global models"""
        print("\n🔍 Comparing Cluster-Based vs Global Models")
        print("="*60)

        # Calculate average metrics for cluster models
        cluster_avg_metrics = {}
        for model_type in ['RandomForest', 'GradientBoosting']:
            cluster_metrics = []
            for cluster_id, results in self.cluster_results.items():
                if model_type in results:
                    cluster_metrics.append(results[model_type]['test']['rmse'])

            if cluster_metrics:
                cluster_avg_metrics[model_type] = np.mean(cluster_metrics)

        # Compare with global models
        print(f"{'Model':<20} {'Global RMSE':<15} {'Cluster Avg RMSE':<20} {'Improvement':<15}")
        print("-"*60)

        for model_type in ['RandomForest', 'GradientBoosting']:
            if model_type in global_results and model_type in cluster_avg_metrics:
                global_rmse = global_results[model_type]['test']['rmse']
                cluster_rmse = cluster_avg_metrics[model_type]
                improvement = (global_rmse - cluster_rmse) / global_rmse * 100

                print(f"{model_type:<20} {global_rmse:<15.4f} {cluster_rmse:<20.4f} {improvement:<15.2f}%")

# ==============================================================================
# VISUALIZATION MODULE
# ==============================================================================

class Visualizer:
    """Class for creating advanced visualizations"""

    @staticmethod
    def plot_consumption_patterns(df, cluster_df):
        """Plot consumption patterns by cluster"""
        fig = make_subplots(
            rows=2, cols=2,
            subplot_titles=('Daily Consumption Pattern', 'Weekly Pattern',
                          'Cluster Distribution', 'Peak Hours by Cluster')
        )

        # Daily pattern
        daily_pattern = df.groupby(df['datetime'].dt.hour)['Global_active_power'].mean()
        fig.add_trace(
            go.Scatter(x=daily_pattern.index, y=daily_pattern.values, name='Hourly Avg'),
            row=1, col=1
        )

        # Weekly pattern
        weekly_pattern = df.groupby(df['datetime'].dt.dayofweek)['Global_active_power'].mean()
        days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
        fig.add_trace(
            go.Bar(x=days, y=weekly_pattern.values, name='Daily Avg'),
            row=1, col=2
        )

        # Cluster distribution
        if 'cluster_kmeans' in cluster_df.columns:
            cluster_counts = cluster_df['cluster_kmeans'].value_counts()
            fig.add_trace(
                go.Pie(labels=cluster_counts.index, values=cluster_counts.values, name='Clusters'),
                row=2, col=1
            )

        # Peak hours by cluster
        if 'cluster_kmeans' in cluster_df.columns and 'peak_hour' in cluster_df.columns:
            peak_hours = cluster_df.groupby('cluster_kmeans')['peak_hour'].mean()
            fig.add_trace(
                go.Bar(x=peak_hours.index.astype(str), y=peak_hours.values, name='Peak Hour'),
                row=2, col=2
            )

        fig.update_layout(height=800, showlegend=True, title_text="Energy Consumption Analysis")
        fig.show()

    @staticmethod
    def plot_predictions(y_true, y_pred, model_name):
        """Plot actual vs predicted values"""
        fig = go.Figure()

        # Sample for better visualization
        sample_size = min(500, len(y_true))
        indices = np.random.choice(len(y_true), sample_size, replace=False)
        indices.sort()

        fig.add_trace(go.Scatter(
            x=indices, y=y_true[indices],
            mode='lines', name='Actual',
            line=dict(color='blue', width=2)
        ))

        fig.add_trace(go.Scatter(
            x=indices, y=y_pred[indices],
            mode='lines', name='Predicted',
            line=dict(color='red', width=2, dash='dash')
        ))

        fig.update_layout(
            title=f'{model_name} - Actual vs Predicted',
            xaxis_title='Sample Index',
            yaxis_title='Power Consumption',
            height=400
        )

        fig.show()

# ==============================================================================
# MAIN EXECUTION PIPELINE
# ==============================================================================

def main():
    """Main execution function"""
    print("\n" + "="*80)
    print(" ENERGY CONSUMPTION FORECASTING WITH PATTERN DISCOVERY ".center(80))
    print("="*80)

    # Step 1: Load Data
    print("\n📁 STEP 1: DATA LOADING")
    print("-"*40)
    loader = DataLoader()
    df_raw = loader.load_data()
    print(f"Dataset shape: {df_raw.shape}")
    print(f"Columns: {list(df_raw.columns)}")

    # Step 2: Preprocess Data
    print("\n🔧 STEP 2: DATA PREPROCESSING")
    print("-"*40)
    preprocessor = DataPreprocessor(df_raw)
    df_clean = preprocessor.preprocess()

    # Step 3: Feature Engineering
    print("\n⚡ STEP 3: FEATURE ENGINEERING")
    print("-"*40)
    fe = FeatureEngineer(df_clean)
    df_features = fe.create_temporal_features()
    daily_features = fe.create_aggregated_features()

    # Step 4: Clustering Analysis
    print("\n🎯 STEP 4: CLUSTERING ANALYSIS")
    print("-"*40)
    clustering = ClusteringAnalysis(daily_features)

    # Find optimal clusters
    optimal_k = clustering.find_optimal_clusters(max_k=6)

    # Perform K-Means clustering
    clustering.perform_kmeans_clustering(n_clusters=optimal_k)

    # Perform Hierarchical clustering
    clustering.perform_hierarchical_clustering(n_clusters=optimal_k)

    # Visualize clusters
    clustering.visualize_clusters()

    # Step 5: Global Forecasting Models
    print("\n📈 STEP 5: GLOBAL FORECASTING MODELS")
    print("-"*40)
    global_forecasting = ForecastingModels(df_features, daily_features)

    # Prepare data
    global_forecasting.prepare_forecasting_data()
    global_forecasting.split_data()

    # Train models
    global_forecasting.train_random_forest()
    global_forecasting.train_gradient_boosting()

    # Compare models
    comparison_df = global_forecasting.compare_models()

    # Step 6: Cluster-Based Forecasting
    print("\n🎯 STEP 6: CLUSTER-BASED FORECASTING")
    print("-"*40)
    cluster_forecasting = ClusterBasedForecasting(df_features, clustering.features_df)
    cluster_forecasting.train_cluster_models()

    # Compare cluster-based vs global models
    cluster_forecasting.compare_cluster_vs_global(global_forecasting.results)

    # Step 7: Visualization
    print("\n📊 STEP 7: FINAL VISUALIZATIONS")
    print("-"*40)
    viz = Visualizer()

    # Plot consumption patterns
    viz.plot_consumption_patterns(df_features, clustering.features_df)

    # Plot predictions for best model
    best_model = 'RandomForest'  # Based on typical performance
    if best_model in global_forecasting.results:
        viz.plot_predictions(
            global_forecasting.y_test,
            global_forecasting.results[best_model]['predictions'],
            best_model
        )

    # Final Summary
    print("\n" + "="*80)
    print(" ANALYSIS COMPLETE ".center(80))
    print(
```

```
# ─── CELL 1: Install Dependencies & Download Dataset ───────────────────────
import subprocess, sys

packages = ['pandas', 'numpy', 'matplotlib', 'seaborn', 'scikit-learn', 'requests']
for pkg in packages:
    subprocess.check_call([sys.executable, '-m', 'pip', 'install', pkg, '-q'])
print('✅ All packages installed!')

import os, urllib.request, zipfile

DATA_DIR  = 'data'
DATA_FILE = os.path.join(DATA_DIR, 'household_power_consumption.txt')
UCI_URL   = 'https://archive.ics.uci.edu/ml/machine-learning-databases/00235/household_power_consumption.zip'

os.makedirs(DATA_DIR, exist_ok=True)

if not os.path.exists(DATA_FILE):
    print('📥 Downloading UCI dataset (~127 MB)...')
    try:
        urllib.request.urlretrieve(UCI_URL, 'temp_data.zip')
        with zipfile.ZipFile('temp_data.zip', 'r') as z:
            z.extractall(DATA_DIR)
        os.remove('temp_data.zip')
        print('✅ Downloaded successfully!')
    except Exception as e:
        print(f'⚠️ Download failed: {e}')
        DATA_FILE = None
else:
    print('✅ Dataset already exists!')
    # ─── CELL 2: Synthetic Data Fallback (if download failed) ──────────────────
import pandas as pd
import numpy as np

SYNTHETIC_FILE = os.path.join(DATA_DIR, 'household_power_consumption_synthetic.csv')

def generate_synthetic_data(n_days=365*3, seed=42):
    np.random.seed(seed)
    dates = pd.date_range(start='2007-01-01', periods=n_days * 24 * 60, freq='T')
    n     = len(dates)
    hour  = dates.hour
    month = dates.month
    dow   = dates.dayofweek

    daily_pattern  = 0.5 * np.sin((hour - 6) * np.pi / 12) + 1.0
    daily_pattern  = np.clip(daily_pattern, 0.2, 2.5)
    seasonal       = 0.3 * np.sin((month - 1) * 2 * np.pi / 12) + 1.0
    weekend_boost  = np.where(dow >= 5, 1.1, 1.0)
    noise          = np.random.normal(0, 0.15, n)
    power          = np.clip(1.0 * daily_pattern * seasonal * weekend_boost + noise, 0.1, 10.0)
    voltage        = 240 + np.random.normal(0, 1.5, n)

    df = pd.DataFrame({
        'Date':                  dates.strftime('%d/%m/%Y'),
        'Time':                  dates.strftime('%H:%M:%S'),
        'Global_active_power':   np.round(power, 3),
        'Global_reactive_power': np.round(power * 0.1 + np.random.normal(0, 0.02, n), 3),
        'Voltage':               np.round(voltage, 3),
        'Global_intensity':      np.round(power * 1000 / voltage, 3),
        'Sub_metering_1':        np.round(np.clip(power * 0.30 + np.random.normal(0, 0.05, n), 0, None), 3),
        'Sub_metering_2':        np.round(np.clip(power * 0.25 + np.random.normal(0, 0.05, n), 0, None), 3),
        'Sub_metering_3':        np.round(np.clip(power * 0.20 + np.random.normal(0, 0.05, n), 0, None), 3),
    })
    return df

if not os.path.exists(DATA_FILE or ''):
    print('🔨 Generating synthetic dataset (3 years, minute-level)...')
    syn_df    = generate_synthetic_data()
    syn_df.to_csv(SYNTHETIC_FILE, sep=';', index=False)
    DATA_FILE = SYNTHETIC_FILE
    print(f'✅ Synthetic dataset saved → {SYNTHETIC_FILE}  shape={syn_df.shape}')
else:
    print(f'✅ Using downloaded data: {DATA_FILE}')
    # ─── CELL 3: Load & Preprocess ─────────────────────────────────────────────
import warnings
warnings.filterwarnings('ignore')

df_raw = pd.read_csv(DATA_FILE, sep=';', na_values=['?', '', 'NA'], low_memory=False)
print(f'✅ Raw data: {df_raw.shape[0]:,} rows × {df_raw.shape[1]} cols')

# Parse datetime index
df_raw['Datetime'] = pd.to_datetime(
    df_raw['Date'] + ' ' + df_raw['Time'], format='%d/%m/%Y %H:%M:%S', errors='coerce')
df_raw.set_index('Datetime', inplace=True)
df_raw.drop(columns=['Date', 'Time'], inplace=True)

# Convert to numeric
numeric_cols = ['Global_active_power','Global_reactive_power','Voltage',
                'Global_intensity','Sub_metering_1','Sub_metering_2','Sub_metering_3']
df_raw[numeric_cols] = df_raw[numeric_cols].apply(pd.to_numeric, errors='coerce')

# Impute
missing = df_raw.isnull().sum()
print('Missing before imputation:', missing[missing > 0].to_dict() or 'None ✅')
df_raw.fillna(method='ffill', inplace=True)
df_raw.fillna(df_raw.mean(numeric_only=True), inplace=True)

# Resample hourly
df_hourly = df_raw.resample('H').mean().dropna()
print(f'✅ Hourly data: {df_hourly.shape[0]:,} rows')
print(f'   Range: {df_hourly.index.min()} → {df_hourly.index.max()}')
df_hourly.head()
# ─── CELL 4: Exploratory Data Analysis ─────────────────────────────────────
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import seaborn as sns

plt.style.use('seaborn-v0_8-whitegrid')
PALETTE = ['#2196F3','#FF5722','#4CAF50','#9C27B0','#FF9800']

fig, axes = plt.subplots(2, 2, figsize=(18, 10))
fig.suptitle('⚡ Household Power Consumption — EDA', fontsize=16, fontweight='bold')

# 1. Time-series (first 30 days)
ax = axes[0, 0]
s  = df_hourly['Global_active_power'].iloc[:30*24]
ax.plot(s.index, s.values, color=PALETTE[0], linewidth=0.8)
ax.set_title('Global Active Power — First 30 Days', fontweight='bold')
ax.set_xlabel('Date'); ax.set_ylabel('Power (kW)')
ax.xaxis.set_major_formatter(mdates.DateFormatter('%b %d'))
plt.setp(ax.xaxis.get_majorticklabels(), rotation=30)

# 2. Average by hour
ax = axes[0, 1]
h_avg = df_hourly.groupby(df_hourly.index.hour)['Global_active_power'].mean()
bars  = ax.bar(h_avg.index, h_avg.values, color=PALETTE[1], edgecolor='white')
peak  = h_avg.idxmax()
bars[peak].set_color(PALETTE[3])
ax.annotate(f'Peak: {peak}:00', xy=(peak, h_avg[peak]),
            xytext=(peak+1, h_avg[peak]*1.02), fontsize=9, color=PALETTE[3], fontweight='bold')
ax.set_title('Average Consumption by Hour', fontweight='bold')
ax.set_xlabel('Hour'); ax.set_ylabel('Avg Power (kW)'); ax.set_xticks(range(0,24,2))

# 3. Monthly averages
ax = axes[1, 0]
m_avg   = df_hourly.groupby(df_hourly.index.month)['Global_active_power'].mean()
m_names = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
ax.bar(range(1,13), [m_avg.get(m,0) for m in range(1,13)], color=PALETTE[2], edgecolor='white')
ax.set_xticks(range(1,13)); ax.set_xticklabels(m_names)
ax.set_title('Average Consumption by Month', fontweight='bold')
ax.set_xlabel('Month'); ax.set_ylabel('Avg Power (kW)')

# 4. Weekday vs Weekend
ax = axes[1, 1]
df_hourly['is_weekend'] = df_hourly.index.dayofweek >= 5
wd = df_hourly[~df_hourly['is_weekend']]['Global_active_power']
we = df_hourly[ df_hourly['is_weekend']]['Global_active_power']
ax.hist(wd, bins=60, alpha=0.65, color=PALETTE[0], label=f'Weekday (n={len(wd):,})', density=True)
ax.hist(we, bins=60, alpha=0.65, color=PALETTE[1], label=f'Weekend (n={len(we):,})', density=True)
ax.axvline(wd.mean(), color=PALETTE[0], linestyle='--', linewidth=1.5)
ax.axvline(we.mean(), color=PALETTE[1], linestyle='--', linewidth=1.5)
ax.set_title('Weekday vs Weekend Distribution', fontweight='bold')
ax.set_xlabel('Power (kW)'); ax.set_ylabel('Density'); ax.legend()

plt.tight_layout()
plt.savefig('eda_overview.png', dpi=150, bbox_inches='tight')
plt.show()

# Correlation heatmap
fig, axes = plt.subplots(1, 2, figsize=(16, 5))
corr = df_hourly[numeric_cols].corr()
mask = np.triu(np.ones_like(corr, dtype=bool))
sns.heatmap(corr, mask=mask, annot=True, fmt='.2f', cmap='coolwarm', center=0,
            ax=axes[0], linewidths=0.5)
axes[0].set_title('Feature Correlation Matrix', fontweight='bold')
plt.setp(axes[0].xaxis.get_majorticklabels(), rotation=45, ha='right')

df_hourly['Day'] = df_hourly.index.day_name().str[:3]
day_order = ['Mon','Tue','Wed','Thu','Fri','Sat','Sun']
sns.boxplot(data=df_hourly[df_hourly['Day'].isin(day_order)],
            x='Day', y='Global_active_power', order=day_order, ax=axes[1], palette='Blues')
axes[1].set_title('Power by Day of Week', fontweight='bold')
plt.tight_layout(); plt.savefig('eda_correlation.png', dpi=150, bbox_inches='tight'); plt.show()
# ─── CELL 5: Feature Engineering ───────────────────────────────────────────
df_daily = df_hourly.resample('D').agg(
    daily_avg_power   = ('Global_active_power','mean'),
    daily_max_power   = ('Global_active_power','max'),
    daily_min_power   = ('Global_active_power','min'),
    daily_std_power   = ('Global_active_power','std'),
    daily_total_power = ('Global_active_power','sum'),
    avg_voltage       = ('Voltage','mean'),
    avg_intensity     = ('Global_intensity','mean'),
    avg_sub1          = ('Sub_metering_1','mean'),
    avg_sub2          = ('Sub_metering_2','mean'),
    avg_sub3          = ('Sub_metering_3','mean'),
).dropna()

# Time features
df_daily['month']        = df_daily.index.month
df_daily['day_of_week']  = df_daily.index.dayofweek
df_daily['is_weekend']   = (df_daily.index.dayofweek >= 5).astype(int)
df_daily['quarter']      = df_daily.index.quarter
df_daily['week_of_year'] = df_daily.index.isocalendar().week.astype(int)

# Peak-period averages
for label, hours in [('morning',range(6,10)),('afternoon',range(12,18)),
                     ('evening',range(18,23)),('night',list(range(0,6))+[23])]:
    mask = df_hourly.index.hour.isin(hours)
    df_daily[f'{label}_avg'] = df_hourly[mask]['Global_active_power'].resample('D').mean()

# Lag & rolling features
for lag, name in [(1,'lag_1d'),(7,'lag_7d'),(30,'lag_30d')]:
    df_daily[name] = df_daily['daily_avg_power'].shift(lag)
df_daily['rolling_7d_avg']  = df_daily['daily_avg_power'].rolling(7,  min_periods=1).mean().shift(1)
df_daily['rolling_30d_avg'] = df_daily['daily_avg_power'].rolling(30, min_periods=1).mean().shift(1)

df_daily.dropna(inplace=True)
print(f'✅ Feature matrix: {df_daily.shape}')
print(f'   Features: {list(df_daily.columns)}')
# ─── CELL 6: K-Means Clustering ────────────────────────────────────────────
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, davies_bouldin_score

CLUSTER_FEATURES = ['daily_avg_power','daily_max_power','daily_std_power',
                    'morning_avg','afternoon_avg','evening_avg','night_avg',
                    'avg_sub1','avg_sub2','avg_sub3','is_weekend']

X_cluster        = df_daily[CLUSTER_FEATURES].copy()
scaler_cluster   = StandardScaler()
X_cluster_scaled = scaler_cluster.fit_transform(X_cluster)

# Elbow + Silhouette
K_range     = range(2, 10)
inertias, silhouettes, db_scores = [], [], []
print('Computing optimal K...')
for k in K_range:
    km     = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = km.fit_predict(X_cluster_scaled)
    inertias.append(km.inertia_)
    silhouettes.append(silhouette_score(X_cluster_scaled, labels))
    db_scores.append(davies_bouldin_score(X_cluster_scaled, labels))
    print(f'  K={k}: Inertia={km.inertia_:.1f} | Silhouette={silhouettes[-1]:.4f}')

best_k = list(K_range)[np.argmax(silhouettes)]
print(f'\n🏆 Optimal K = {best_k}')

# Plot evaluation curves
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
fig.suptitle('🔵 K-Means Cluster Evaluation', fontsize=14, fontweight='bold')
K_list = list(K_range)
for ax, (vals, title, color) in zip(axes, [
    (inertias,   'Elbow (Inertia)',                   'bo-'),
    (silhouettes,'Silhouette Score (Higher = Better)', 'go-'),
    (db_scores,  'Davies-Bouldin (Lower = Better)',   'ro-'),
]):
    ax.plot(K_list, vals, color, linewidth=2, markersize=8)
    ax.axvline(best_k, color='red', linestyle='--', alpha=0.7, label=f'Best K={best_k}')
    ax.set_title(title, fontweight='bold'); ax.set_xlabel('K'); ax.legend()
plt.tight_layout(); plt.savefig('cluster_evaluation.png', dpi=150, bbox_inches='tight'); plt.show()

# Fit final model
N_CLUSTERS         = best_k
kmeans             = KMeans(n_clusters=N_CLUSTERS, random_state=42, n_init=10)
df_daily['cluster']= kmeans.fit_predict(X_cluster_scaled)

cluster_counts = df_daily['cluster'].value_counts().sort_index()
print('Cluster distribution:')
for c, cnt in cluster_counts.items():
    print(f'  Cluster {c}: {cnt} days ({cnt/len(df_daily)*100:.1f}%)')

# Cluster profiles heatmap
cluster_profile = df_daily.groupby('cluster')[CLUSTER_FEATURES].mean().round(4)
profile_norm = (cluster_profile - cluster_profile.min()) / (cluster_profile.max() - cluster_profile.min())

plt.figure(figsize=(10, 6))
sns.heatmap(profile_norm.T, cmap='YlGnBu', annot=True, fmt='.2f', linewidths=.5)
plt.title('Cluster Profiles (Normalized)', fontweight='bold')
plt.xlabel('Cluster')
plt.ylabel('Feature')
plt.savefig('cluster_profiles.png', dpi=150, bbox_inches='tight')
plt.show()

# ─── CELL 7: Train All Models ───────────────────────────────────────────────
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

FEATURE_COLS = [c for c in [
    'month','day_of_week','is_weekend','quarter','week_of_year',
    'daily_max_power','daily_min_power','daily_std_power',
    'avg_voltage','avg_intensity','avg_sub1','avg_sub2','avg_sub3',
    'morning_avg','afternoon_avg','evening_avg','night_avg',
    'lag_1d','lag_7d','lag_30d','rolling_7d_avg','rolling_30d_avg'
] if c in df_daily.columns]
TARGET_COL = 'daily_avg_power'

X = df_daily[FEATURE_COLS]
y = df_daily[TARGET_COL]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)
scaler    = StandardScaler()
X_train_s = scaler.fit_transform(X_train)
X_test_s  = scaler.transform(X_test)

print(f'✅ Train: {len(X_train)} | Test: {len(X_test)} | Features: {len(FEATURE_COLS)}')

def evaluate_model(name, y_true, y_pred):
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    mae  = mean_absolute_error(y_true, y_pred)
    r2   = r2_score(y_true, y_pred)
    print(f'  {name:45s} | RMSE={rmse:.5f} | MAE={mae:.5f} | R²={r2:.5f}')
    return {'Model': name, 'RMSE': rmse, 'MAE': mae, 'R2': r2}

results = []

# ── GLOBAL MODELS ──────────────────────────────────────────────────────────
print('\n' + '='*70)
print('  GLOBAL MODELS')
print('='*70)

rf_global = RandomForestRegressor(n_estimators=200, max_depth=15,
                                   min_samples_split=5, random_state=42, n_jobs=-1)
rf_global.fit(X_train_s, y_train)
rf_pred = rf_global.predict(X_test_s)
results.append(evaluate_model('Global — Random Forest', y_test, rf_pred))

gb_global = GradientBoostingRegressor(n_estimators=200, learning_rate=0.05,
                                       max_depth=5, subsample=0.8, random_state=42)
gb_global.fit(X_train_s, y_train)
gb_pred = gb_global.predict(X_test_s)
results.append(evaluate_model('Global — Gradient Boosting', y_test, gb_pred))

# ── CLUSTER-BASED MODELS ───────────────────────────────────────────────────
print('\n' + '='*70)
print('  CLUSTER-BASED MODELS')
print('='*70)

df_train_full = df_daily.iloc[:len(X_train)]
df_test_full  = df_daily.iloc[len(X_train):]
rf_c_preds    = pd.Series(index=df_test_full.index, dtype=float)
gb_c_preds    = pd.Series(index=df_test_full.index, dtype=float)

labels_map = {0: 'Low Consumption', 1: 'High Consumption'}

for c in sorted(df_daily['cluster'].unique()):
    train_idx = df_train_full[df_train_full['cluster'] == c].index
    test_idx  = df_test_full[df_test_full['cluster'] == c].index
    if len(train_idx) < 20 or len(test_idx) < 5:
        print(f'  ⚠️ Cluster {c}: insufficient data, skipping.')
        continue
    Xc_tr = X.loc[train_idx];  yc_tr = y.loc[train_idx]
    Xc_te = X.loc[test_idx];   yc_te = y.loc[test_idx]
    sc    = StandardScaler()
    Xc_tr_s = sc.fit_transform(Xc_tr)
    Xc_te_s = sc.transform(Xc_te)

    rf_c = RandomForestRegressor(n_estimators=150, max_depth=10, random_state=42, n_jobs=-1)
    rf_c.fit(Xc_tr_s, yc_tr)
    rf_c_preds.loc[test_idx] = rf_c.predict(Xc_te_s)

    gb_c = GradientBoostingRegressor(n_estimators=150, learning_rate=0.05, max_depth=4, random_state=42)
    gb_c.fit(Xc_tr_s, yc_tr)
    gb_c_preds.loc[test_idx] = gb_c.predict(Xc_te_s)

    lbl = labels_map.get(c, f'Cluster {c}')
    print(f'\n  [{lbl}]  train={len(train_idx)}  test={len(test_idx)}')
    evaluate_model(f'  C{c} — Random Forest', yc_te, rf_c.predict(Xc_te_s))
    evaluate_model(f'  C{c} — Gradient Boosting', yc_te, gb_c.predict(Xc_te_s))

valid = rf_c_preds.notna() & gb_c_preds.notna()
y_tv  = y_test[valid]
print('\n' + '='*70)
print('  CLUSTER AGGREGATED RESULTS')
print('='*70)
results.append(evaluate_model('Cluster-Based — Random Forest (All)',    y_tv, rf_c_preds[valid]))
results.append(evaluate_model('Cluster-Based — Gradient Boosting (All)',y_tv, gb_c_preds[valid]))
# ─── CELL 8: Evaluate & Visualize ──────────────────────────────────────────
df_results = pd.DataFrame(results).sort_values('RMSE').reset_index(drop=True)
df_results[['RMSE','MAE','R2']] = df_results[['RMSE','MAE','R2']].round(5)

print('\n' + '='*70)
print('  FINAL MODEL COMPARISON')
print('='*70)
print(df_results.to_string(index=False))
best = df_results.iloc[0]
print(f'\n🏆 Best Model: {best["Model"]}')
print(f'   RMSE={best["RMSE"]:.5f}  MAE={best["MAE"]:.5f}  R²={best["R2"]:.5f}')

# ── Bar chart comparison ───────────────────────────────────────────────────
fig, axes = plt.subplots(1, 3, figsize=(18, 6))
fig.suptitle('📊 Model Performance Comparison', fontsize=14, fontweight='bold')
bar_colors = ['#42A5F5' if 'Cluster' in m else '#EF5350' for m in df_results['Model']]
for ax, metric in zip(axes, ['RMSE','MAE','R2']):
    bars = ax.barh(range(len(df_results)), df_results[metric],
                   color=bar_colors, edgecolor='white', height=0.6)
    ax.set_yticks(range(len(df_results)))
    ax.set_yticklabels([m[:40] for m in df_results['Model']], fontsize=8)
    ax.set_title(metric, fontweight='bold'); ax.set_xlabel(metric)
    for bar, val in zip(bars, df_results[metric]):
        ax.text(val + 0.0001, bar.get_y() + bar.get_height()/2, f'{val:.4f}', va='center', fontsize=7.5)
from matplotlib.patches import Patch
fig.legend(handles=[Patch(facecolor='#EF5350', label='Global'),
                    Patch(facecolor='#42A5F5', label='Cluster-Based')],
           loc='lower center', ncol=2, bbox_to_anchor=(0.5,-0.02))
plt.tight_layout(); plt.savefig('model_comparison.png', dpi=150, bbox_inches='tight'); plt.show()

# ── Actual vs Predicted scatter ────────────────────────────────────────────
fig, axes = plt.subplots(2, 2, figsize=(16, 12))
fig.suptitle('🔍 Actual vs Predicted', fontsize=14, fontweight='bold')
for ax, (name, pred, actual) in zip(axes.flat, [
    ('Global — RF',  rf_pred, y_test.values),
    ('Global — GB',  gb_pred, y_test.values),
    ('Cluster — RF', rf_c_preds[valid].values, y_tv.values),
    ('Cluster — GB', gb_c_preds[valid].values, y_tv.values),
]):
    ax.scatter(actual, pred, alpha=0.3, s=15, color='#2196F3')
    lims = [min(actual.min(), pred.min()), max(actual.max(), pred.max())]
    ax.plot(lims, lims, 'r--', linewidth=1.5)
    ax.set_title(f'{name}  R²={r2_score(actual, pred):.4f}  RMSE={np.sqrt(mean_squared_error(actual,pred)):.4f}',
                 fontsize=10, fontweight='bold')
    ax.set_xlabel('Actual (kW)'); ax.set_ylabel('Predicted (kW)')
plt.tight_layout(); plt.savefig('actual_vs_predicted.png', dpi=150, bbox_inches='tight'); plt.show()

# ── Feature importance ──────────────────────────────────────────────────────
fig, axes = plt.subplots(1, 2, figsize=(18, 7))
fig.suptitle('🔎 Feature Importance', fontsize=14, fontweight='bold')
for ax, (model, name, color) in zip(axes, [
    (rf_global,'Random Forest','#42A5F5'),
    (gb_global,'Gradient Boosting','#FF7043')
]):
    imp = pd.Series(model.feature_importances_, index=FEATURE_COLS).sort_values().tail(15)
    ax.barh(imp.index, imp.values, color=color, edgecolor='white')
    ax.set_title(f'{name} — Top 15 Features', fontweight='bold'); ax.set_xlabel('Importance')
plt.tight_layout(); plt.savefig('feature_importance.png', dpi=150, bbox_inches='tight'); plt.show()

# ── Forecast time-series ───────────────────────────────────────────────────
test_dates = df_daily.iloc[len(X_train):].index
n_show     = min(90, len(test_dates))
fig, ax    = plt.subplots(figsize=(18, 6))
ax.plot(test_dates[:n_show], y_test.values[:n_show], label='Actual', color='black', linewidth=1.5)
ax.plot(test_dates[:n_show], rf_pred[:n_show], label='RF Global', color='#2196F3', linestyle='--')
ax.plot(test_dates[:n_show], gb_pred[:n_show], label='GB Global', color='#FF5722', linestyle=':')
ax.plot(test_dates[:n_show], rf_c_preds[valid].reindex(test_dates[:n_show]),
        label='RF Cluster', color='#4CAF50', linestyle='--')
ax.plot(test_dates[:n_show], gb_c_preds[valid].reindex(test_dates[:n_show]),
        label='GB Cluster', color='#9C27B0', linestyle=':')
ax.set_title('Energy Consumption Forecast', fontweight='bold')
ax.set_xlabel('Date')
ax.set_ylabel('Power (kW)')
ax.legend()
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('forecast_timeseries.png', dpi=150, bbox_inches='tight')
plt.show()


# ─── CELL 9: Final Summary ──────────────────────────────────────────────────
print('='*70)
print('  PROJECT SUMMARY')
print('='*70)
print(f'  Hourly records  : {len(df_hourly):,}')
print(f'  Daily vectors   : {len(df_daily):,}')
print(f'  Features        : {len(FEATURE_COLS)}')
print(f'  Optimal K       : {N_CLUSTERS}')
print(f'\n  Cluster Labels  :')
for k, v in labels_map.items():
    cnt = (df_daily['cluster'] == k).sum()
    print(f'    Cluster {k} → {v}  ({cnt} days)')
print(f'\n  Model Results (sorted by RMSE):')
print(df_results[['Model','RMSE','MAE','R2']].to_string(index=False))
g_rf  = df_results[df_results['Model'].str.contains('Global.*Random')]['RMSE'].values
c_rf  = df_results[df_results['Model'].str.contains('Cluster.*Random.*All')]['RMSE'].values
if len(g_rf) and len(c_rf):
    delta = (g_rf[0] - c_rf[0]) / g_rf[0] * 100
    sign  = '✅ IMPROVED' if delta > 0 else '⚠️  No improvement'
    print(f'\n  Clustering Impact on RF RMSE: {sign} by {abs(delta):.2f}%')
print(f'\n  Saved plots: eda_overview.png, cluster_profiles.png,')
print(f'               model_comparison.png, forecast_timeseries.png')
print('='*70)
print('✅ All done!')
```
